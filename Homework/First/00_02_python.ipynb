```python
# тут ваш код и примеры его работы

class Matrix(object):
    # two-dimentions matrix, store columns
    
    def __init__(self,matrix):
        self.cells = matrix
    
    def add(self, other):
        if isinstance(other,Matrix) and len(self.cells[0]) == len(other.cells):
            result = []
            for i in range(0,len(self.cells)):
                if len(self.cells[i]) == len(other.cells[i]):
                    column = []
                    for j in range(0,len(self.cells[i])):
                        column.append(self.cells[i][j] + other.cells[i][j])
                    result.append(column)    
                else:
                    return None
            return Matrix(matrix = result)    
        return None            
                        
            
    def mul(self,coef):
        result = []
        for column in self.cells:
            new_column = []
            for elem in column:
                new_column.append(elem * coef)
            result.append(new_column)    
        return Matrix(matrix = result)    
        
    def mul_matrix(self,other):
        if isinstance(other,Matrix) and len(self.cells) == len(other.cells):
            result = []
            for i in range(0,len(self.cells)):
                if len(self.cells[i]) == len(other.cells[i]):
                    column = []
                    for j in range(0,len(self.cells[i])):
                        elem = 0
                        for k in range (0,len(self.cells)):
                            elem += self.cells[i][k] * other.cells[k][j]
                        column.append(elem)
                    result.append(column)    
                else:
                    return None
            return Matrix(matrix = result)    
        return None  
        
    def determinant(self):
        
        if len(self.cells) != len(self.cells[0]):
            return None
        
        if len(self.cells) <= 1:
            return None
        
        if len(self.cells) == 2:
            return self.cells[0][0] * self.cells[1][1] - self.cells[1][0] * self.cells[0][1]
        
        column = self.cells[0]
        result = 0
        for k in range(0,len(column)):
            new_matrix = []
            for i in range(1,len(self.cells)):
                new_matrix.append([])
                for j in range(0,len(self.cells[i])):
                    if (j != k):
                        new_matrix[i-1].append(self.cells[i][j])
            temp_matrix = Matrix(matrix = new_matrix)     
            result += column[k] * ((-1)**k) * temp_matrix.determinant()
        return result
    
    def reversed_matrix(self):
        if (self.determinant() == 0 or self.determinant() is None):
            return None
        
        if (len(self.cells) <= 1) :
            return None
        
        if (len(self.cells) == 2) :
            minor_matrix = Matrix(matrix = [[self.cells[1][1],(-1) * 
            self.cells[1][0]],[self.cells[0][0],(-1) * self.cells[0][1]]])
            return minor_matrix.transponned(),self.determinant
        
        minor = []
        for i in range(0,len(self.cells)):
            minor.append([])
            for j in range(0,len(self.cells[i])):
                determ_matrix = []
                counter = 0
                for k in range(0,len(self.cells)):
                    if (k == i):
                        continue
                    determ_matrix.append([])    
                    for l in range(0,len(self.cells[i])):
                        if (l == j):
                            continue
                        determ_matrix[counter].append(self.cells[k][l])
                    counter+=1
                minor[i].append(Matrix(matrix = determ_matrix).determinant())    
        
        for i in range(0,len(minor)):
            for j in range(0,len(minor[i])):
                minor[i][j] *= (-1)**(i + j) 
                    
        minor_matrix = Matrix(matrix = minor)    
        
        return minor_matrix.transponned(),self.determinant()
    
    def transponned(self):
        result = []
        for i in range(0,len(self.cells)):
            result.append([])
        for i in range(0,len(self.cells)):
            for j in range(0,len(self.cells[i])):
                    result[j].append(self.cells[i][j])          
        return Matrix(matrix = result)    
    
matrix = Matrix(matrix = [[1,2,3],[4,5,6],[7,8,9]])
matrix_2 = Matrix(matrix = [[1,2,1],[1,1,3],[13,1,1]])
print(matrix.add(matrix_2).cells)
print(matrix.mul(3).cells)
print(matrix.mul_matrix(matrix_2).cells)
print(matrix_2.determinant())
reverse = matrix_2.reversed_matrix()
print(reverse[0].cells,reverse[1])
print(matrix_2.mul_matrix(reverse[0]).mul(1/reverse[1]).cells)
print(matrix.transponned().cells)
        
        
```
